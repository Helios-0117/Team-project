### Data Analyzing 資料分析

```python
starting_fund = 1000000
aggressive = [0.3  ,0.15,0    ,0.1  ,0.15,0   ,0.15 ,0   ,0.15]
conservative = [0.075,0   ,0.075,0.075,0   ,0.35,0.075,0.35,0   ]
steady = [0.15 ,0.10,0.15 ,0.15 ,0   ,0.15,0.15 ,0.15,0   ]
A_fund = []
C_fund = []
S_fund = []
for i in range (0,9):
    A_fund.append(starting_fund * aggressive[i])
    C_fund.append(starting_fund * conservative[i])
    S_fund.append(starting_fund * steady[i])
A_portfolio = []
C_portfolio = []
S_portfolio = []
for j in range (1,10):
    A_portfolio.append(int(A_fund[j-1] / float(final.iloc[0,j-1])))
    C_portfolio.append(int(C_fund[j-1] / float(final.iloc[0,j-1])))
    S_portfolio.append(int(S_fund[j-1] / float(final.iloc[0,j-1])))
A_net_worth = []
C_net_worth = []
S_net_worth = []
length = final.shape[0]
for k in range (length):
    A_net = 0
    C_net = 0
    S_net = 0
    for kk in range (1,10):
        A_net = A_net + int(A_portfolio[kk-1] * float(final.iloc[k,kk-1]))
        C_net = C_net + int(C_portfolio[kk-1] * float(final.iloc[k,kk-1]))
        S_net = S_net + int(S_portfolio[kk-1] * float(final.iloc[k,kk-1]))
    A_net_worth.append(A_net)
    C_net_worth.append(C_net)
    S_net_worth.append(S_net)
dict_A = {}
dict_C = {}
dict_S = {}
mytime = final.index.array
for m in range(0,length):
    dict_A[mytime[m]] = A_net_worth[m]

df_A = pd.DataFrame.from_dict(dict_A, orient='index', columns=['積極型 即時淨值'])
for m in range(0,length):
    dict_C[mytime[m]] = C_net_worth[m]

df_C = pd.DataFrame.from_dict(dict_C, orient='index', columns=['保守型 即時淨值'])
for m in range(0,length):
    dict_S[mytime[m]] = S_net_worth[m]

df_S = pd.DataFrame.from_dict(dict_S, orient='index', columns=['穩健型 即時淨值'])

temp08 = final.join(df_A)
temp09 = temp08.join(df_C)
finale = temp09.join(df_S)
# finale.to_csv('Organized0827.csv')
```

```python
Aggressive_cash_flow = []
Conservative_cash_flow = []
Steady_cash_flow = []
Aggressive_cash = []
Conservative_cash = []
Steady_cash = []
total_A = total_C = total_S = 0
tport_A = [3490, 8064, 0, 0, 6576, 0, 8125] 
tport_C = [872, 0, 2780, 9485, 3288, 18087, 0] 
tport_S = [1745, 5376, 5561, 4065, 6576, 7751, 0]

for p in range (yearly_cash_flow.shape[0]):
    tA = tC = tS = 0
    for q in range (yearly_cash_flow.shape[1]):
        tA = tA + tport_A[q] * float(yearly_cash_flow.iloc[p,q])
        tC = tC + tport_C[q] * float(yearly_cash_flow.iloc[p,q])
        tS = tS + tport_S[q] * float(yearly_cash_flow.iloc[p,q])
    Aggressive_cash.append(tA)
    Conservative_cash.append(tC)
    Steady_cash.append(tS)
    Aggressive_cash_flow.append('$' + str(int(tA) // 1000) + ',' + str(int(tA))[2:])
    Conservative_cash_flow.append('$' + str(int(tC) // 1000) + ',' + str(int(tC))[2:])
    Steady_cash_flow.append('$' + str(int(tS) // 1000) + ',' + str(int(tS))[2:])
for r in range (len(Aggressive_cash)):
    total_A = total_A + Aggressive_cash[r]
for s in range (len(Conservative_cash)):
    total_C = total_C + Conservative_cash[s]
for t in range (len(Steady_cash)):
    total_S = total_S + Steady_cash[t]
year.append('Total')
Aggressive_cash_flow.append('$' + str(int(total_A) // 1000) + ',' + str(int(total_A))[3:])
Conservative_cash_flow.append('$' + str(int(total_C) // 1000) + ',' + str(int(total_C))[3:])
Steady_cash_flow.append('$' + str(int(total_S) // 1000) + ',' + str(int(total_S))[3:])
```
